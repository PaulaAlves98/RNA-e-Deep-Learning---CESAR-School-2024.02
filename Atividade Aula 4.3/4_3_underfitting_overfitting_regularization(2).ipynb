número de camadas

número de neurônios em cada camada

Configuração Bem-Sucedida (Exemplo Alcançado):

Features: X1, X2

Camadas Ocultas: 3 camadas (7 neurônios na primeira, 5 na segunda, 3 na terceira)

Ativação: ReLU (em todas as camadas ocultas)

Learning Rate: 0.1

Regularização: L2

Fator de Regularização: 0.001

Resultado: Esta configuração levou a um Test loss de 0.033 e um Training loss de 0.010 após cerca de 1062 épocas. A fronteira de decisão resultante separou as espirais de forma eficaz, e a pequena diferença entre o loss de treino e teste indicou um bom controle do overfitting.
